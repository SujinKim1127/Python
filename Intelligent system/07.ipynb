```python
class info():
    def __init__(self, name, num, score=[]):
        self.name = name
        self.num = num
        self.score = score
    def average(self):
        sum = 0
        for i in range(0,2):
            sum = sum + self.score[i]
        avg = sum / len(self.score)
        print(avg)
    def __repr__(self):
        print(self.name)
        print(self.num)
        return "{}".format(self.score)

name = input()
num = input()
score=[]
for a in range (0,2):
    s = int(input())
    score.append(s)

student=info(name, num, score)
print(student)
```
