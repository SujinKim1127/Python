## 문제3 
위와 같은 ‘how_many_a’ list가 있을 때, Attribute Function을 사용하여 [결과] 와 같이 출력하는 프로그램을 작성하시오.
(반드시 for문, if문, Attribute Function 사용)

```python
how_many_a = [{'a':False}, "aaaabbbccd", 23, ["a","a","a","d","s"]]
for element in how_many_a:
    if hasattr(element, "count"):
        print(element, end = "  ")
        print(element.count("a"))
```

## 문제4 
위와 같은 class 가 있을 때, arg 변수를 출력하니 “Hong” 이 아닌 Instance object 가 출력된다. Instance object 가 아닌 “Hong” 이 출력되도록 새로운 Dunder method를 추가하여 해결하시오.

```python
class Student():
    def __init__(self, argument_name):
        self.name = argument_name
    def __repr__(self):
        return self.name

arg = Student("Hong")
print(arg)
```

## 문제5
5-1) class 와 Dunder method 를 사용하여 본인이름, 본인학번, 임의의 점수(2개)를 입력받는 프로그램을 작성하고, [출력예시] 와 같이 이름과 학번과 점수 2개를 출력하시오. 


```python
class info():
    def __init__(self, name, num, score=[]):
        self.name = name
        self.num = num
        self.score = score
    def average(self):
        sum = 0
        for i in range(0,2):
            sum = sum + self.score[i]
        avg = sum / len(self.score)
        print(avg)
    def __repr__(self):
        print(self.name)
        print(self.num)
        return "{}".format(self.score)

name = input()
num = input()
score=[]
for a in range (0,2):
    s = int(input())
    score.append(s)

student=info(name, num, score)
print(student)
```

            (위의 class 안에 “average” 함수 정의)
