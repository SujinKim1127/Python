## 문제1
결과 1과 같은 출력 결과가 나오도록 빈칸을 채우시오.

```python
array = np.arange(24).reshape(3,4,2)
array
```

```python
s0 = array.sum(axis = 0)
s1 = array.sum(axis = 2)
s2 = array.sum(axis = 1)
print(s0)
print()
print(s1)
print()
print(s2)
```

```python
s0 = array.mean(axis = 2)
s1 = array.mean(axis = 1)
s2 = array.mean(axis = 0)
print(s0)
print()
print(s1)
print()
print(s2)
```
